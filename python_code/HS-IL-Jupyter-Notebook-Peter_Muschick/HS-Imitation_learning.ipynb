# Locally Weighted Projection Regression (LWPR)

Locally Weighted Projection Regression (LWPR) uses nonlinear function approximation in high dimensional spaces with redundant  and irrelevant input dimensions. Locally linear models are used to be spanned by a small number of univariate regressions in specific input space directions. A locally weighted variant of Partial Least Squares (PLS) is employed for doing the dimensionality reduction.

```python
import numpy as np
import matplotlib.pyplot as plt
```

# Genereting data

```python
# range of sin
range = 30
# random scatter factor
min_max = 0.2

x = np.arange(range-range, range, 0.01)
delta = np.random.uniform(-min_max, min_max, size=(3000,))

y = np.sin(x) + delta

plt.plot(x, y)
plt.show()
```

# Generating data for imitation

Two baseline algorithms were used to generate data for imitation learning purpose.

## NeuroEvolution Algorithm

Based on [NeuroEvolution Algorithm Git repo](https://github.com/SurenderHarsha/OpenAIGym-Cartpole-Evolution-/) learning data was generated. Initially 120k lines where used. After 100 successfull (over 200 timesteps balancing) epochs, 600 epochs were recorded and printed into a file, which was used to learn the algorithm. 

A few problems occured:
* Short initial phase to balance out the beginning setup of the pole
